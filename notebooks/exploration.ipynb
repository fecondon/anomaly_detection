```python
from tensorflow.keras.models import load_model, Model
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
```

```python
model = load_model('models/autoencoder.h5')
encoder = Model(inputs=model.input, outputs=model.layers[3].output)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.drop(columns=['is_anomaly', 'reconstruction_error', 'predicted_anomaly']))
latent = encoder.predict(X_scaled)
```

```python
latent_2d = PCA(n_components=2).fit_transform(latent)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(latent_2d[:, 0], latent_2d[:, 1],
            c=df['reconstruction_error'], cmap='coolwarm', alpha=0.7)
plt.title('Latent Space Colored by Reconstruction Error')
plt.xlabel('Latent Dim 1')
plt.ylabel('Latent Dim 2')
plt.colorbar(label='Reconstruction Error')
plt.show()
```
